```python
static_path = '../static'
PCR_config = {
    'seq_len': 500,
    'mut_pad': 5,
    'prod_size_min': 120,
    'prod_size_max': 220
}
filter1_p3 = run_primer3(p_df, chroms_folder=f"{static_path}/genome/gatk/hg38/split", pcr_config=PCR_config,  primer_list='')
filter1_p3
```

```python
filter1_newprimers = filter1_df.loc[:,:'NormalHDRinfo'].merge(filter1_p3.drop(columns=['Chr','Start','End','Ref','Alt','Gene','offsetR','AmpliconRange', 'InsertSeq']).rename(columns={'Note':'Temp', 'Status':'note'}), left_index=True, right_index=True)
filter1_newprimers
```
